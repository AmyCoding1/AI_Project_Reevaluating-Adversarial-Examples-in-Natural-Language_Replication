```python
import textattack
import transformers
import time

# Define models and tokenizers
bert_model = transformers.AutoModelForSequenceClassification.from_pretrained("textattack/bert-base-uncased-imdb")
lstm_model = textattack.models.helpers.LSTMForClassification()

bert_tokenizer = transformers.AutoTokenizer.from_pretrained("textattack/bert-base-uncased-imdb")
# Use a space-based tokenizer for the LSTM model
lstm_tokenizer = textattack.models.tokenizers.SpaceTokenizer()

# Define model wrappers
bert_model_wrapper = textattack.models.wrappers.HuggingFaceModelWrapper(bert_model, bert_tokenizer)
lstm_model_wrapper = textattack.models.wrappers.HuggingFaceModelWrapper(lstm_model, lstm_tokenizer)

# Define attack recipes
bert_attack = textattack.attack_recipes.TextFoolerJin2019.build(bert_model_wrapper)
lstm_attack = textattack.attack_recipes.GeneticAlgorithm(building_blocks=["swap", "delete", "insert"])

from textattack.attack_methods import Attack
from textattack.constraints.pre_transformation import (
    RepeatModification,
    StopwordModification,
)
from textattack.shared import AttackedText
from textattack.transformations import CompositeTransformation, Swap, Delete, Insert

class LSTMGeneticAttack(Attack):
    def __init__(self, transformation, constraints=None):
        transformation = CompositeTransformation([transformation])
        constraints = (
            [RepeatModification(), StopwordModification()] + (constraints or [])
        )
        super().__init__(constraints=constraints, transformation=transformation)

# Define LSTM attack recipe
lstm_attack = LSTMGeneticAttack(transformation=Swap(), constraints=[RepeatModification()])


# Define datasets
imdb_dataset = textattack.datasets.HuggingFaceDataset("imdb", split="test")
yelp_dataset = textattack.datasets.HuggingFaceDataset("yelp_polarity", split="test")
movie_review_dataset = textattack.datasets.HuggingFaceDataset("rotten_tomatoes", split="test")

# Run TextFooler with BERT on IMDb, Yelp Polarity, and Movie Review datasets
for dataset, dataset_name in zip([imdb_dataset, yelp_dataset, movie_review_dataset], ["IMDb", "Yelp Polarity", "Movie Review"]):
    attack_args_bert = textattack.AttackArgs(
        num_examples=2,  # Adjust the number of examples as needed
        log_to_csv=f"/content/drive/MyDrive/Project_AI/log_bert_{dataset_name.lower()}.csv",
        disable_stdout=True
    )
    attacker_bert = textattack.Attacker(bert_attack, dataset, attack_args_bert)
    attacker_bert.attack_dataset()

# Run GeneticAttack with LSTM on IMDb, Yelp Polarity, and Movie Review datasets
for dataset, dataset_name in zip([imdb_dataset, yelp_dataset, movie_review_dataset], ["IMDb", "Yelp Polarity", "Movie Review"]):
    attack_args_lstm = textattack.AttackArgs(
        num_examples=2,  # Adjust the number of examples as needed
        log_to_csv=f"/content/drive/MyDrive/Project_AI/log_lstm_{dataset_name.lower()}.csv",
        disable_stdout=True
    )
    attacker_lstm = textattack.Attacker(lstm_attack, dataset, attack_args_lstm)
    attacker_lstm.attack_dataset()

    # Start the timer
start_time = time.time()

# Run the attack
attacker.attack_dataset()

# End the timer
end_time = time.time()

# Calculate and print the total running time
running_time = end_time - start_time
print(f"The attack took {running_time} seconds to run.")
```
