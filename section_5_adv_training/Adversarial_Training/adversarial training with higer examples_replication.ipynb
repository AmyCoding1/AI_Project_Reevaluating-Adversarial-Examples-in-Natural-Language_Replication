```python
import textattack
import transformers
import time

# Define BERT model and tokenizer
bert_model = transformers.AutoModelForSequenceClassification.from_pretrained("textattack/bert-base-uncased-imdb")
bert_tokenizer = transformers.AutoTokenizer.from_pretrained("textattack/bert-base-uncased-imdb")

# Define BERT model wrapper
bert_model_wrapper = textattack.models.wrappers.HuggingFaceModelWrapper(bert_model, bert_tokenizer)

# Define TextFooler attack
textfooler_attack = textattack.attack_recipes.TextFoolerJin2019.build(bert_model_wrapper)

# Define datasets
movie_review_dataset = textattack.datasets.HuggingFaceDataset("rotten_tomatoes", split="test")

# Run TextFooler with BERT on IMDb, Yelp Polarity, and Movie Review datasets
for dataset, dataset_name in zip([movie_review_dataset], ["Movie Review"]):
    attack_args = textattack.AttackArgs(
        num_examples=100,  # Adjust the number of examples as needed
        log_to_csv=f"/content/drive/MyDrive/Project_AI/log_textfooler_bert_{dataset_name.lower()}.csv",
        disable_stdout=True
    )
    attacker = textattack.Attacker(textfooler_attack, dataset, attack_args)
    attacker.attack_dataset()

# Start the timer
start_time = time.time()

# Run the attack
attacker.attack_dataset()

# End the timer
end_time = time.time()

# Calculate and print the total running time
running_time = end_time - start_time
print(f"The attack took {running_time} seconds to run.")
```

```python
import textattack
from textattack.attack_recipes import AttackRecipe
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# Load attack recipe from file
attack_recipe_path = '/content/drive/MyDrive/Project_AI/Reevaluating-NLP-Adversarial-Examples/section_6_adjusted_attacks/recipes/textfooler_jin_2019_adjusted.py'
with open(attack_recipe_path, 'r') as file:
    attack_code = file.read()

# Load pre-trained Hugging Face model and tokenizer
model_name = './outputs/2023-12-16-17-29-24-499210/best_model/'  # Replace with the actual model name you are using
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Create a model wrapper with the loaded model and tokenizer
model_wrapper = textattack.models.wrappers.HuggingFaceModelWrapper(model, tokenizer)

# Create an attack recipe class dynamically from the code and build it with the model wrapper
exec(attack_code)
attack_recipe_class = [c for c in locals().values() if isinstance(c, type) and issubclass(c, AttackRecipe)][0]
textfooler_attack = attack_recipe_class.build(model_wrapper)

# Define datasets
movie_review_dataset = textattack.datasets.HuggingFaceDataset("rotten_tomatoes", split="test")

# Run the loaded attack on Movie Review dataset for 10 epochs
for dataset, dataset_name in zip([movie_review_dataset], ["Movie Review"]):
    attack_args = textattack.AttackArgs(
        num_examples=10,  # Adjust the number of examples as needed
        num_epochs=10,   # Set the number of epochs
        log_to_csv=f"/content/drive/MyDrive/Project_AI/tfadjusted_attack_{dataset_name.lower()}_10_epochs.csv",
        disable_stdout=True
    )
    attacker = textattack.Attacker(textfooler_attack, dataset, attack_args)
    attacker.attack_dataset()
```
