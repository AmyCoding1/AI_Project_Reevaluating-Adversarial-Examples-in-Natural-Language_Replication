```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Project_AI/Replication-NLP-Adversarial-Examples-master/section_5_adv_training/non_suspicious_results.csv')
```

```python
df['Input.original'].unique()
```

```python
def fix(x):
    return 'Real' if x else 'Computer-altered'

df['Orig'] = df['Input.original'].apply(fix)
```

```python
df['Correct'] = df['Orig'] == df['Answer.sentiment.label']
len(df[df['Correct']])
```

```python
df_corr = df[df['Correct']]
```

```python
df_sus_corr = df_corr[df_corr['Orig'] == 'Computer-altered']
len(df_sus_corr)
```
